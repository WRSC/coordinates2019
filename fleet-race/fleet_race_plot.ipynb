```python
import folium
import pandas
import yaml
```

```python
trajectory = pandas.read_csv('../fleet-race/south/HUST 1-Fleet Race-26-1-1.csv',
                             names=['time','latitude','longitude','speed','course'])
```

```python
marker_list = [['A', [29.86713941703848, 121.5389755240182]],
 ['B', [29.86705406261338, 121.5385997804418]],
 ['C', [29.86720864100981, 121.5386780355552]],
 ['D', [29.86729537733915, 121.5384019484949]]]
```

```python
def plot_track(file_name):
    trajectory = pandas.read_csv(file_name, names=['time','latitude','longitude','speed','course'])
    map_osm = folium.Map(location=[29.86705406261338, 121.5385997804418], zoom_start=17)
    
    for name, latlon in marker_list:
            folium.CircleMarker(latlon, radius=5, popup=name).add_to(map_osm)
    latlons = [(row.latitude, row.longitude) for row in trajectory.itertuples()]
    map_osm.add_child(folium.PolyLine(latlons))
    return map_osm
```

```python
plot_track('../fleet-race/south/HUST 1-Fleet Race-26-1-1.csv')
```
