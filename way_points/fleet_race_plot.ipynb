```python
import folium
import pandas
import glob
from IPython.display import HTML, display
```

```python
south_marker_list = [['A', [29.86713941703848, 121.5389755240182]],
 ['B', [29.86705406261338, 121.5385997804418]],
 ['C', [29.86720864100981, 121.5386780355552]],
 ['D', [29.86729537733915, 121.5384019484949]]]
```

```python
def plot_track(file_name, marker_list):
    trajectory = pandas.read_csv(file_name, names=['time','latitude','longitude','speed','course'])
    map_osm = folium.Map(location=[29.86705406261338, 121.5385997804418], zoom_start=17)
    
    for name, latlon in marker_list:
            folium.CircleMarker(latlon, radius=5, popup=name).add_to(map_osm)
    latlons = [(row.latitude, row.longitude) for row in trajectory.itertuples()]
    map_osm.add_child(folium.PolyLine(latlons))
    return map_osm
```

```python
south_teams_attempts_am = glob.glob('../fleet-race/south/AM/*.csv')
display(HTML('<h1>Morning Attempts -- NORTH</h1>'))
for south_teams_attempt in south_teams_attempts_am:
    attempt_name = south_teams_attempt.split('/')[-1]
    display(HTML('<h3>{}</h3>'.format(attempt_name)))
    display(plot_track(south_teams_attempt, south_marker_list))
```

```python
south_teams_attempts_pm = glob.glob('../fleet-race/south/PM/*.csv')
display(HTML('<h1>Afternoon Attempts -- NORTH</h1>'))
for south_teams_attempt in south_teams_attempts_pm:
    attempt_name = south_teams_attempt.split('/')[-1]
    display(HTML('<h3>{}</h3>'.format(attempt_name)))
    display(plot_track(south_teams_attempt, south_marker_list))
```
